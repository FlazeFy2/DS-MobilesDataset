```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Exploratory Data Analysis (EDA) - Line Chart Total Device Released per Year for Each Merk
# Dataset
df = pd.read_csv("Mobiles Dataset (2025).csv", encoding="ISO-8859-1")

grouped = df.groupby(['Launched Year', 'Company Name']).size().reset_index(name='Total')
pivot_df = grouped.pivot(index='Launched Year', columns='Company Name', values='Total').fillna(0)

# Plot the Chart
plt.figure(figsize=(12, 6))
for company in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[company], marker='o', label=company)
# Line Chart
plt.title('Total Devices Released per Year by Merk')
plt.xlabel('Launch Year')
plt.ylabel('Number of Devices')
plt.legend(title='Company Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Exploratory Data Analysis (EDA) - Line Chart Average Battery Capacity for Device Released per Year for Each Merk
# Dataset
df = pd.read_csv("Mobiles Dataset (2025).csv", encoding="ISO-8859-1")

df['Battery Capacity'] = df['Battery Capacity'].str.replace('mAh', '').str.replace(',', '').str.strip()
df['Battery Capacity'] = pd.to_numeric(df['Battery Capacity'], errors='coerce')
grouped = df.groupby(['Launched Year', 'Company Name'])['Battery Capacity'].mean().reset_index()

# Pivot 
pivot_df = grouped.pivot(index='Launched Year', columns='Company Name', values='Battery Capacity').fillna(0)

# Plot the chart
plt.figure(figsize=(12, 6))
for company in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[company], marker='o', label=company)

# Line chart
plt.title('Average Battery Capacity per Year by Merk')
plt.xlabel('Launch Year')
plt.ylabel('Battery Capacity (mAh)')
plt.legend(title='Company Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Exploratory Data Analysis (EDA) - Line Chart Average Device RAM for Device Released per Year for Each Merk
# Dataset
df = pd.read_csv("Mobiles Dataset (2025).csv", encoding="ISO-8859-1")

df['RAM'] = df['RAM'].str.replace('GB', '').str.replace(',', '').str.strip()
df['RAM'] = pd.to_numeric(df['RAM'], errors='coerce')
grouped = df.groupby(['Launched Year', 'Company Name'])['RAM'].mean().reset_index()

# Pivot 
pivot_df = grouped.pivot(index='Launched Year', columns='Company Name', values='RAM').fillna(0)

# Plot the chart
plt.figure(figsize=(12, 6))
for company in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[company], marker='o', label=company)

# Line chart
plt.title('Average Device RAM per Year by Merk')
plt.xlabel('Launch Year')
plt.ylabel('RAM (GB)')
plt.legend(title='Company Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
