```python
import pandas as pd
import statistics
import numpy as np
import matplotlib.pyplot as plt
import squarify

# Declaration
# Exploratory Data Analysis (EDA) - Tree Map Total Mobile Device per Context
target_col = ['Company Name','Front Camera']

for dt in target_col:
    data = pd.read_csv("Mobiles Dataset (2025).csv", encoding="ISO-8859-1")
    
    total_context = data.groupby(dt)['Model Name'].count().reset_index()
    
    # Sort 
    total_context = total_context.sort_values(by='Model Name', ascending=False)
    
    # Extract
    sizes = total_context['Model Name']
    labels = total_context.apply(lambda x: f"{x[dt]}\n({x['Model Name']})", axis=1)
    
    # Plot Tree Map
    plt.figure(figsize=(15, 9))
    squarify.plot(sizes=sizes, label=labels, alpha=0.8)
    plt.title(f"Number of Mobile Device per {dt}", fontsize=14)
    plt.axis('off')
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import squarify

# Exploratory Data Analysis (EDA) - Tree Map The Most Expensive Device per Merk
data = pd.read_csv("Mobiles Dataset (2025).csv", encoding="ISO-8859-1")

# Clean the Data
data['Launched Price (USA)'] = (
    data['Launched Price (USA)'].str.replace('USD', '', regex=False).str.replace(',', '', regex=False).str.strip().astype(float)
)

# Find the Max Value
idx = data.groupby('Company Name')['Launched Price (USA)'].idxmax()
result = data.loc[idx, ['Company Name', 'Model Name', 'Launched Price (USA)']]
result = result.reset_index(drop=True)

# Plot Tree Map
sizes = result['Launched Price (USA)']
labels = result.apply(lambda x: f"{x['Company Name']}\n{x['Model Name']}\n${x['Launched Price (USA)']:.0f}", axis=1)
plt.figure(figsize=(15, 12))
squarify.plot(sizes=sizes, label=labels, alpha=0.8)
plt.title("Most Expensive Mobile Device per Merk", fontsize=14)
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import squarify

# Exploratory Data Analysis (EDA) - Tree Map The Most Highest Battery Capacity per Merk
data = pd.read_csv("Mobiles Dataset (2025).csv", encoding="ISO-8859-1")

# Clean the Data
data['Battery Capacity'] = (
    data['Battery Capacity'].str.replace('mAh', '', regex=False).str.replace(',', '', regex=False).str.strip().astype(float)
)

# Find the Max Value
idx = data.groupby('Company Name')['Battery Capacity'].idxmax()
result = data.loc[idx, ['Company Name', 'Model Name', 'Battery Capacity']]
result = result.reset_index(drop=True)

# Plot Tree Map
sizes = result['Battery Capacity']
labels = result.apply(lambda x: f"{x['Company Name']}\n{x['Model Name']}\n{x['Battery Capacity']:.0f}mAh", axis=1)
plt.figure(figsize=(15, 12))
squarify.plot(sizes=sizes, label=labels, alpha=0.8)
plt.title("Most Highest Battery Capacity per Merk", fontsize=14)
plt.axis('off')
plt.show()
```
